```python
from sympy import *
init_printing()
```

```python
x = symbols("x")
f = x**2-8*x-20
solve(f,x) # equation roots
```

```python
f # pretty print the equation itself
```

```python
# plot the graph
plot(f)
```

```python
diff(f,x) # print the derivative
```

```python
diff(f,x).subs(x,5) # evaluate at x=5
```

```python
# plot both functions
p = plot(f, diff(f, x), show=False)
p[1].line_color = 'r'
p.show()
```
